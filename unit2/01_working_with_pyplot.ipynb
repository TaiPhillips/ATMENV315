# Working with pyplot


```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
L=50.0
x=np.arange(0,101,1)
sindata=np.sin(2*np.pi*x/L)
cosdata=np.cos(2*np.pi*x/L)


plt.plot(x,sindata,color='red')
plt.plot(x,cosdata,color='green')
plt.show()
```
