# Working with pyplot


```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
L=50.0
x=np.arange(0,101,1)
sindata=np.sin(2*np.pi*x/L)
cosdata=np.cos(2*np.pi*x/L)


plt.plot(x,sindata,color='red')
plt.plot(x,cosdata,color='green')
plt.show()
```

### <span style="color:blue">Problem 1:</span> 
We have two functions in this plot. For people with indifferent perception of red and green color tones we want change the color from green to blue.


### <span style="color:darkgreen">Solution 1:</span> 

We can create a list or 1-dimensional numpy array that contains index position numbers. Then, we use the index-list (numpy array) inside the square brackets of the variables x, sindata, and cosdata.

```python
ihelp=np.arange(5,90,3)

plt.plot(x,sindata,color='red')
plt.plot(x,cosdata,color='blue')
plt.show()
```

### <span style="color:blue">Problem 2:</span> 
Add a legend to the figure.


### <span style="color:darkgreen">Solution 2:</span> 

```python
plt.plot(x,sindata,color='red',label='sin(x)')
plt.plot(x,cosdata,color='blue',label='cos(x)')
plt.legend()
plt.show()
```

### <span style="color:blue">Problem 3:</span> 
Change this graph into a bar graph, with the colors in gold, and purple, with a legend.

```python
x=[1,2,3,4]
y=[1,4,9,16]
z=[-16,-8,0,8]
```

```python
plt.plot(x,y)
plt.plot(x,z)
```

### <span style="color:darkgreen">Solution 3:</span> 


```python
plt.bar(x,y,color='gold',label="data 1")
plt.bar(x,z,color='purple',label="data 2")
plt.legend()
plt.show()
```

### <span style="color:blue">Problem 4:</span> 

Add axis labels to the barplot: 
- x-axis: "year"
- y-axis: "win/loss [%]"

### <span style="color:darkgreen">Solution 4:</span> 

```python
plt.bar(x,y,color='gold',label="data 1")
plt.bar(x,z,color='purple',label="data 2")
plt.xlabel("year")
plt.ylabel("win/loss [%]")
plt.legend()
plt.show()
```

### <span style="color:darkgreen">Solution 5:</span> 

Last but not least: use the function plt.scatter to create a scatter plot of the same data. Add axis labels and a legend as before.

```python
plt.scatter(x,y,color='gold',label="data 1")
plt.scatter(x,z,color='purple',label="data 2")
plt.xlabel("year")
plt.ylabel("win/loss [%]")
plt.legend()
plt.show()
```
